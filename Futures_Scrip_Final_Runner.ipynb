```python
import os
import openpyxl
# package import statement
from SmartApi import SmartConnect #or from smartapi.smartConnect import SmartConnect
#import smartapi.smartExceptions(for smartExceptions)
import pandas
from datetime import datetime, timedelta


#create object of call
obj=SmartConnect(api_key="UmjRRBor")

#login api call

data = obj.generateSession("R51647002","1797","272940")
refreshToken= data['data']['refreshToken']

#fetch the feedtoken
# feedToken=obj.getfeedToken()
print(data)

def generate_dates(start_date):
    dates = []
    current_date = datetime.strptime(start_date, '%Y-%m-%d')

    for _ in range(365):
        dates.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)

    return dates


# Example usage
start_date = '2022-06-28'
year_dates = generate_dates(start_date)
```

```python
import time
# Load the workbook
workbook = openpyxl.load_workbook('filterred_scripts_type.xlsx')

# Select the active sheet
sheet = workbook.active

# Define the column indices for the relevant columns
symbol_name_column_index = 4
symbol_type_column_index = 9
symbol_token_column_index = 8


# Iterate through the rows starting from the second row (skipping the header)
for row in sheet.iter_rows(min_row=2):
    # Get the symbol name and symbol token
    symbol_name = row[symbol_name_column_index].value
    symbol_type = row[symbol_type_column_index].value
    symbol_token = row[symbol_token_column_index].value

    # print(symbol_name + ":" + symbol_type + ":" + symbol_token)
    print(symbol_name)
    print(symbol_type)
    print(symbol_token)

    # Create the folder path
    folder_path = os.path.join(symbol_name, symbol_type)

    # Check if the folder already exists
    if not os.path.exists(folder_path):
        # Create the folder
        os.makedirs(folder_path)
        print(folder_path)

    for date in year_dates:
        time.sleep(0.2)
        print(symbol_name + ":" + symbol_type + "/" + "data_"+ date + ".csv")
        #Historic api
        try:
            historicParam={
            "exchange": "NFO",
            "symboltoken": symbol_token,
            "interval": "ONE_MINUTE",
            "fromdate": date + " 09:15",
            "todate": date + " 15:30"
            }
            data = obj.getCandleData(historicParam)
            pandas.DataFrame(data["data"]).to_csv(symbol_name+"/"+symbol_type+"/"+ date + ".csv")
        # print(data)
        except Exception as e:
            print("Historic Api failed: {}".format(e.message))

# Save the updated workbook
workbook.save('filterred_scripts_type.xlsx')
```
